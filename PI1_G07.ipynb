```python
#Importo las librerias necesarias
import pandas as pd
```

```python
#Abro el csv y muestro el encabezado para corroborar que se haya subido correctamente
df = pd.read_csv("csv\\netflix.csv",delimiter = ',',encoding = "utf-8")
df.head()
```

```python
#Creo una copia auxiliar del csv utilizando .copy() para trabajar sin modificar al original.
df_aux = df.copy()
df_aux.head()
```

```python
#Cambio el tipo de la columna ["date_added"] a datetime utilizando to_datetime()
df["date_added"] = pd.to_datetime(df["date_added"])
df.head()
```

```python
#Ordeno de mas antiguo a mas reciente usando : ".sort_values(by=['date_added'])"
df = df.sort_values(by=["date_added"])
df.head()
```

```python
#Reemplazo en la columna "director" el valor "Not Given" por None utilizando .replace
df["director"] = df["director"].replace({"Not Given":None})
df.head()
```

```python
#Utilizo máscaras para dividir 3 dataframes distintos según la columna "release_year": 2019, 2020, 2021

df2019 = df.release_year == 2019
df2019 = df[df2019]
df2019.head()
```

```python
df2020 = df.release_year == 2020
df2020 = df[df2020]
df2020.head()
```

```python
df2021 = df.release_year == 2021
df2021 = df[df2021]
df2021.head()
```

```python
#Transformo los 3 dataframes en 3 diccionarios distintos usando: .reset_index().to_dict(orient="index")

df2019dicc = df2019.reset_index().to_dict(orient="index")
df2020dicc = df2020.reset_index().to_dict(orient="index")
df2021dicc = df2021.reset_index().to_dict(orient="index")
```

```python
import pandas as pd

df = pd.read_csv("csv\\netflix.csv",delimiter = ',',encoding = "utf-8")

df["date_added"] = pd.to_datetime(df["date_added"])

df = df.sort_values(by=["date_added"])

df["director"] = df["director"].replace({"Not Given":None})

df2019 = df.release_year == 2019
df2019 = df[df2019]

df2020 = df.release_year == 2020
df2020 = df[df2020]

df2021 = df.release_year == 2021
df2021 = df[df2021]

df2019dicc = df2019.reset_index().to_dict(orient="index")
df2020dicc = df2020.reset_index().to_dict(orient="index")
df2021dicc = df2021.reset_index().to_dict(orient="index")
```
